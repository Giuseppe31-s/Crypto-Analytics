## Taxa de Retorno Simples

A taxa de retorno simples é uma medida básica usada para calcular o desempenho de um investimento ao longo de um período de tempo específico. Ela é expressa como uma porcentagem e representa o ganho ou perda percentual em relação ao investimento inicial.

Para calcular a taxa de retorno simples, você precisa das seguintes informações:

1. Valor inicial do investimento (ou custo inicial).
2. Valor final do investimento (ou valor atual).
3. Período de tempo para o qual o retorno está sendo calculado.

A fórmula para calcular a taxa de retorno simples é:

$${Taxa de retorno simples} = \frac{{Valor final do investimento} - {Valor inicial do investimento}}{{Valor inicial do investimento}} \times 100\%$$

Por exemplo, suponha que você tenha investido R$ 10.000 em ações e, ao final de um ano, o valor do seu investimento tenha crescido para R$ 11.500. Para calcular a taxa de retorno simples, você usaria a fórmula da seguinte maneira:


$${Taxa de retorno simples} = \frac{11.500 - 10.000}{10.000} \times 100\% = \frac{1.500}{10.000} \times 100\% = 0,15 \times 100\% = 15\%$$

Portanto, a taxa de retorno simples do seu investimento seria de 15%.

É importante notar que a taxa de retorno simples não leva em consideração o tempo ou qualquer fluxo de caixa intermediário durante o período de investimento. Isso significa que ela pode ser mais adequada para comparar o desempenho de investimentos de curto prazo ou para situações onde não há fluxos de caixa intermediários.
